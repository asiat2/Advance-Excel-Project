```python
def plot_histogram(model_df, column, unit):
    plt.hist(model_df[column], bins=10, edgecolor='black')
    plt.title(f'{column}')
    plt.xlabel(f'{column} ({unit})')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
def plot_normalized_histogram(model_df, column, label):
    model_df[column] = (model_df[column] - model_df[column].mean()) / model_df[column].std()
    plt.hist(model_df[column], bins=10, alpha=0.5, label=label, edgecolor='black')

# Normalize and plot histograms
plt.figure(figsize=(10,6))
plot_normalized_histogram(model_df, 'Patient Satisfaction', 'Patient Satisfaction')
plot_normalized_histogram(model_df, 'Satisfaction Enhancement', 'Satisfaction Enhancement')

plt.title('Enhancement Feedback Outcome Vs. Satisfaction Feedback ')
plt.xlabel('Normalized Values')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the range of intervention values to explore
intervention_values = [0.1 * i for i in range(1, 6)]

# List of interventions
interventions = [

       'Access to Healthcare Improvements Interventions',
       'Education and Employment Support Interventions',
       'Community Health Education Interventions',
       'Community Safety Programs Interventions',
       'Comprehensive Data Collection Interventions'

]

# List of stocks
stocks = [
    'Patient Engagement',
    'Patient Satisfaction',
    'Communication Effectiveness',
    'Patient Health Outcomes'

]

# Loop over each intervention
for intervention in interventions:
    # Loop over each stock
    for stock in stocks:
        plt.figure(figsize=(10, 6))
        # Loop over each intervention value
        for value in intervention_values:
            # Set the intervention value
            model.set_components({intervention: value})

            # Run the model
            output = model.run()

            # Plot the output for the stock
            plt.plot(output.index, output[stock], label=f'Intervention Value: {value}')

        plt.title(f'Effect of {intervention} on {stock}')
        plt.xlabel('Time ($Year$)')
        plt.ylabel('Population')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.grid(True)
        plt.show()
```

```python


# Loading the simulated output data from the PySD
data = pd.read_csv('Patient_Centric_Model.csv')

data = data.rename(columns={'Unnamed: 0': 'Year'})

# Specifying the target variable
target = 'Patient Satisfaction'

# Implementing the features from the data
features = data.columns.drop(target)

# Split the data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], test_size=0.2, random_state=42)

# Initialize the models
models = {
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42)

}

# Train and evaluate the models
for name, model in models.items():
    model.fit(X_train, y_train)  # Train the model
    predictions = model.predict(X_test)  # Make predictions on the test set
    mae = mean_absolute_error(y_test, predictions)
    mse = mean_squared_error(y_test, predictions)
    rmse = np.sqrt(mse)
    print(f"{name}:\nMAE: {mae}\nMSE: {mse}\nRMSE: {rmse}\n")




```
